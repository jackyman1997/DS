# 2. Data cleaning and Exploratory Data Analysis (lazy version)

```python
# data cleaning part 1, removing nan columns
keep = []
for key in data.keys():  # iterating thru columns 
    if not data[key].isna().all():  # if not all the data in that column is Nan
        keep.append(key)  # key that column

data_clean_1 = data[keep]
```

```python
data_clean_1.describe()
```

```python
# data cleaning part 2, removing nan rows
data_clean_2 = data_clean_1.copy()
for i, row in enumerate(data_clean_1.iterrows()):  # iterating rows by rows
    if data_clean_1.iloc[i].isna().any():  # if any of the data in that row is Nan
        data_clean_2.drop(i, axis=0, inplace=True)  # drop that row
```

```python
data_clean_2.describe()
```

```python
# data cleaning part 3, removing useless data with just guessing/common sense
## date is hard and kinda irrelevant here, so kick that off 
## but this can be very important if a time series is assumed
data_clean_3 = data_clean_2.drop('Date', axis=1)
```

```python
# data cleaning part 4, encode them 
data_clean_4 = data_clean_3.copy()
for key in data_clean_4.keys():  # iterating thru columns 
    if data_clean_4[key].dtypes == 'object':  # if the dtype is 'object' (ie. string)
        if key == "FTR":  # column FTR is the label
            y_df = pd.get_dummies(data_clean_4[key])  # so i used one-hot encoder
        else: 
            data_clean_4[key] = data_clean_4[key].astype('category').cat.codes  # catergorise them and replace them as code

X_df = data_clean_4.drop('FTR', axis=1)
```

# 3. Modelling
I m not doing any hyperparameters tuning and standard scaling here for simplicity sake.  
Since it is a multiclass classification problem, I m using hamming loss and accuracy as metrics.  
For longer y (more label dimension), cosine similarity can also be used.  
[read more here](https://stats.stackexchange.com/questions/233275/multilabel-classification-metrics-on-scikit)

Models are selected from [sklearn multiclass page](https://scikit-learn.org/stable/modules/multiclass.html), only considering the models that support multilabel: 
- tree.DecisionTreeClassifier
- tree.ExtraTreeClassifier
- ensemble.ExtraTreesClassifier
- neighbors.KNeighborsClassifier
- neural_network.MLPClassifier
- neighbors.RadiusNeighborsClassifier
- ensemble.RandomForestClassifier
- linear_model.RidgeClassifierCV

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import hamming_loss, accuracy_score  # for multi label classification metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier, RadiusNeighborsClassifier
from sklearn.linear_model import RidgeClassifierCV
from sklearn.ensemble import ExtraTreesClassifier, RandomForestClassifier
from sklearn.neural_network import MLPClassifier
```

```python
# a function for simplicity
def scoreMe(model, truth, predict): 
    print(f'''
        model: {model.__class__.__name__}
        hamming loss: {hamming_loss(truth, predict)}
        accuracy: {accuracy_score(truth, predict)}
    ''')
```

```python
# to numpy
y = y_df.to_numpy()
X = X_df.to_numpy()

# train test split, val is for the final final, dun use it now
X_trn, X_tst, y_trn, y_tst = train_test_split(X, y, test_size=0.25, random_state=42)
X_val, X_tst, y_val, y_tst = train_test_split(X_tst, y_tst, test_size=0.5, random_state=42)
```

```python
models = [
    DecisionTreeClassifier(), 
    KNeighborsClassifier(), 
    # RadiusNeighborsClassifier(), 
    # RidgeClassifierCV(), 
    ExtraTreesClassifier(), 
    RandomForestClassifier(), 
    MLPClassifier()
]

trained = []
for model in models: 
    model.fit(X_trn, y_trn)
    y_pred = model.predict(X_tst)
    scoreMe(model, y_tst, y_pred)
    trained.append(model)
```

___
# 4. Validation
Same as the above but use the trained model on val data

```python
for model in trained: 
    y_pred = model.predict(X_val)
    scoreMe(model, y_val, y_pred)
```
